```python
# Handling missing values
data.ffill(inplace=True)

# Convert 'Date' column to datetime if 'Date' column exists
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
else:
    print("The 'Date' column is not in the dataset.")

# Set 'Date' as the index if 'Date' column exists
if 'Date' in data.columns:
    data.set_index('Date', inplace=True)
else:
    print("The 'Date' column is not in the dataset.")

# Checking for outliers using boxplot
if 'Sales (ZAR)' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data['Sales (ZAR)'])
    plt.title('Sales Data - Outlier Check')
    plt.show()
else:
    print("The 'Sales (ZAR)' column is not in the dataset.")
```

```python
from prophet import Prophet

df_prophet = data[['DateR', 'Sales (ZAR)']].copy()
df_prophet = df_prophet.rename(columns={'DateR': 'ds', 'Sales (ZAR)': 'y'})

# Converting 'ds' column to datetime format
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])

# Checking for missing values
print(df_prophet[['ds', 'y']].isnull().sum())

# Initializing Prophet model
model_prophet = Prophet()

model_prophet.fit(df_prophet)

# Makimg future dataframe for predictions
future = model_prophet.make_future_dataframe(periods=90)
forecast = model_prophet.predict(future)

# Ploting forecast
model_prophet.plot(forecast)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Checking if the data is stationary (using ADF test)
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['Sales (ZAR)'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

# Differencing the data if needed
data_diff = data['Sales (ZAR)'].diff().dropna()

# Fiting the ARIMA model
model_arima = ARIMA(data_diff, order=(5,1,0))
model_arima_fit = model_arima.fit()

forecast_arima = model_arima_fit.forecast(steps=90)

# Ploting ARIMA forecast
plt.figure(figsize=(10,6))
plt.plot(forecast_arima)
plt.title('ARIMA Forecast')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

# Scaling the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data['Sales (ZAR)'].values.reshape(-1, 1))

# Creatimg sequences for training the LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(data_scaled, seq_length)

# Spliting into training and test sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Building the LSTM model
model_lstm = tf.keras.Sequential([
    tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    tf.keras.layers.LSTM(50, return_sequences=False),
    tf.keras.layers.Dense(25),
    tf.keras.layers.Dense(1)
])

model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model_lstm.fit(X_train, y_train, batch_size=64, epochs=10)

# Making predictions
predictions = model_lstm.predict(X_test)
predictions = scaler.inverse_transform(predictions)

# Ploting LSTM predictions
plt.figure(figsize=(10,6))
plt.plot(data.index[-len(predictions):], predictions, label='Predicted')
plt.plot(data.index[-len(predictions):], data['Sales (ZAR)'][-len(predictions):], label='Actual')
plt.legend()
plt.title('LSTM Predictions vs Actual')
plt.show()
```
